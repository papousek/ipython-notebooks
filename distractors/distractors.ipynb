# Algorithm for Construction of Distractors

The following IPython notebook presents ideas to improve an algorithm for question construction used by www.outlinemaps.org.

Firstly we need to load data:
- Flashcards correspond to items practiced by the system, e.g. Czech Republic on the map of Europe. Each flashcards has a unique **item id**.
- For each item we have its difficulty computed by Elo rating system.
- For each pair of items we have a **confusing factor**, number counting wrong answers (systems asks about item A, users answers B => confusing factor for (A, B) increases by 1).

```python
flashcards = pd.read_csv('./data/flashcards.csv')
difficulty = pd.read_csv('./data/difficulty.csv')
confusing_factor = pd.read_csv('./data/confusing_factor.csv')
```

For further analysis we use data only for European countries.

```python
flashcards = flashcards[(flashcards['term_type'] == 'state') & (flashcards['context_name'] == 'Europe')]
difficulty = difficulty[difficulty['item_id'].isin(flashcards['item_id'].unique())]
confusing_factor = confusing_factor[confusing_factor['item_asked_id'].isin(flashcards['item_id'].unique())]
```

It is beneficial to look at the data:

- Difficulty computed by Elo rating system is a real number which can be trasformed to the interval [0, 1] representing probability an average users answers an open question for the given item correctly. **How are these probabilities distributed?**
- The system uses a scoring function making users' practice suitable to their skill. A parameter for this function is **target probability**.
- We can also look at distribution of confusing factor (here for "Serbia")

```python
def sigmoid(x):
    'Transform a real number to interval [0, 1]'
    return 1.0 / (1 + exp(-x))

def score_probability(target_probability, probability):
    'Compute a score when for the given probability when the target probability is given.'
    diff = target_probability - probability
    sign = 1 if diff > 0 else -1
    normed_diff = abs(diff) / max(0.001, abs(target_probability - 0.5 + sign * 0.5))
    return 1 - normed_diff ** 2

def plot_look_data(target_probability=0.75, term_name='Serbia'):
    rcParams['figure.figsize'] = 15, 5
    xs = np.linspace(0, 1, 100)
    plt.subplot(121)
    plt.hist(map(lambda x: sigmoid(-x), difficulty['difficulty']))
    plt.xlabel('Difficulty')
    plt.ylabel('Number of items')
    plt.twinx()
    plt.plot(xs, map(lambda x: score_probability(target_probability, x), xs), color='black', linewidth=5)
    plt.ylabel('Score')
    plt.subplot(122)
    item_id = flashcards[flashcards['term_name'] == term_name]['item_id'].values[0]
    cf = confusing_factor[confusing_factor['item_asked_id'] == item_id].set_index('item_answered_id')['confusing_factor'].to_dict()
    item_ids, cfs = zip(*sorted(cf.items(), key=lambda x: -x[1])[:20])
    flashcard_names = flashcards.set_index('item_id')['term_name'].to_dict()
    plt.plot(cfs, linewidth=2)
    plt.fill_between(range(len(cfs)), [0] * len(cfs), cfs, alpha=0.2)
    plt.xlim(0, len(cfs) - 1)
    plt.ylabel('Confusing factor')
    plt.title('{}'.format(term_name))
    plt.xticks(range(len(cfs)), map(lambda i: flashcard_names[i], item_ids), rotation=90)
    plt.tight_layout()

def demo_data():
    term_name = widgets.SelectWidget(values=sorted(flashcards['term_name'].unique()))
    target_probability = widgets.FloatSliderWidget(min=0, max=1, step=0.01, value=0.75)
    w = widgets.interactive(
        plot_look_data,
        target_probability=target_probability,
        term_name=term_name
    )
    display(w)
demo_data()
```

Based on users' history we usually adjust the target probability. For less successful users we make the practice easier, for users with high skill, the practice is more challenging.

```python
def adjust(target_probability, rolling_success):
    """
    Adjust the practice difficulty when the target probability
    and rolling success (past 10 users' answers) are given.
    """
    if rolling_success is None:
        return target_probability
    norm = 1 - target_probability if rolling_success > target_probability else target_probability
    correction = ((target_probability - rolling_success) / max(0.001, norm)) * (1 - norm)
    return target_probability + correction

def plot_adjustment(target_probability=0.75, rolling_success=0.75):
    rcParams['figure.figsize'] = 15, 5
    xs = np.linspace(0, 1, 100)
    plt.subplot(121)
    plt.plot(xs, map(lambda x: adjust(target_probability, x), xs), linewidth=5)
    plt.axvline(target_probability, 0, target_probability, '.-', color='gray')
    plt.axhline(target_probability, 0, target_probability, '.-', color='gray')
    plt.plot([target_probability], [target_probability], '.', markersize=20, marker='o', color='gray')
    plt.text(0.05, target_probability + 0.025, 'target probability', color='gray')
    plt.text(target_probability - 0.05, 0.325, 'target probability', color='gray', rotation='vertical')
    plt.xlabel('Rolling success')
    plt.ylabel('Adjusted target probability')
    plt.subplot(122)
    adjusted_probability = adjust(target_probability, rolling_success)
    flashcard_names = flashcards.set_index('item_id')['term_name'].to_dict()
    names, difficulties, scores = zip(*sorted([
    (flashcard_names[i], sigmoid(-d), score_probability(adjusted_probability, sigmoid(-d)))
     for (i, d) in difficulty.set_index('item_id')['difficulty'].to_dict().items()], key=lambda x: -x[2])[:20])
    plt.tight_layout()
    plt.plot(scores, linewidth=5, label='Score')
    plt.xticks(range(len(names)), names, rotation=90)
    plt.plot(difficulties, color='gray', label='Difficulty')
    plt.xlim(0, len(scores) - 1)
    plt.legend(loc=2)
    
def demo_adjustment():
    rolling_success = widgets.FloatSliderWidget(min=0, max=1, step=0.1, value=0.7)
    target_probability = widgets.FloatSliderWidget(min=0, max=1, step=0.01, value=0.75)
    w = widgets.interactive(
        plot_adjustment,
        rolling_success=rolling_success,
        target_probability=target_probability
    )
    display(w)
    
demo_adjustment()
```

Difficulty is not the only criterium for which we have a scoring function. There is also a number of a user's answers for the given item and time since the last answer. It can easily happen the chose item does not fit to the target probability. In this case we do not construct an open question, but a multiple-choice one.
- How many distractors should this question have?
- Which items should be selected as distractors?

Now, look at the algorithm to determine the number of distractors. The original algorithm tried to ensure the real difficulty is as close as possible to the target one. To have more varied questions, it seems better to adopt more uniform strategy.

```python
def adjust_distractors_number(target_probability, difficulty, max_options=6):
    round_fun = round
    g = min(0.5, max(0, target_probability - difficulty) / max(0.001, 1 - difficulty))
    k = round_fun(1.0 / g) if g != 0 else 0
    return int(0 if (k > max_options or k == 0) else k)

def uniform_distractors_number(target_probability, difficulty, max_options=6):
    if difficulty >= target_probability:
        return 0
    return 1 + max(int(ceil((difficulty / target_probability) * (max_options - 1))), 1)

def plot_distractors_number(target_probability=0.75):
    rcParams['figure.figsize'] = 15, 5
    for i, strategy in enumerate([uniform_distractors_number, adjust_distractors_number]):
        xs = np.linspace(0, 1, 100)
        plt.subplot(1, 2, i + 1)
        plt.title(strategy.__name__)
        distractors = map(lambda x: strategy(target_probability, x), xs)
        scores = map(lambda x: score_probability(target_probability, x), xs)
        plt.plot(xs,distractors)
        plt.fill_between(xs, [0] * len(distractors), distractors, alpha=0.2)
        plt.xlabel('Difficulty')
        plt.ylabel('Number of options')
        plt.twinx()
        plt.ylabel('Score / Real difficulty')
        plt.plot(xs, scores, color='gray', linewidth=2)
        plt.plot(xs, map(lambda (x, d): (1.0 / d + (1 - 1.0 / d) * x) if d > 0 else x, zip(xs, distractors)), color='black', linewidth=5)
    plt.tight_layout()
    
def demp_distractors_number():
    target_probability = widgets.FloatSliderWidget(min=0, max=1, step=0.01, value=0.75)
    w = widgets.interactive(
        plot_distractors_number,
        target_probability=target_probability
    )
    display(w)
    
demp_distractors_number()
```

After choosing the appropriate number of distractors. We must choose distractors themselves. The original algorithm selected the most competitive items. On the other hand in some situations (e.g. when Czech people try to learn provinces of China), it seems better to adjust the competitivness based on a user's skill.

```python
def get_level(target_probability, difficulty):
    return min(difficulty / target_probability, 1.0)


def adjust_to_level(level, x, op, median):
    print level
    if x > median:
        if level > 0.5:
            result = median + (x - median) * ((level - 0.5) / 0.5)
        else:
            result = op + (median - op) * (level / 0.5)
    else:
        if level > 0.5:
            result = x + (median - x) * ((level - 0.5) / 0.5)
        else:
            result = median + (op - median) * (level / 0.5)
    print level, x, op, median, result
    return result

def roulette(items_dict, n):
    if n > len(items_dict):
        raise Exception("Can't choose {} samples from {} items".format(n, len(items_dict)))
    items = items_dict.items()
    chosen = set()
    for i in range(n):
        total = sum(zip(*items)[1])
        dice = random.random() * total
        running_weight = 0
        chosen_item = None
        for item, weight in items:
            if dice < running_weight + weight:
                chosen_item = item
                break
            running_weight += weight
        chosen.add(chosen_item)
        items = [(i, w) for (i, w) in items if i != chosen_item]
    return chosen


def competive_distractors(target_probability, item_asked, prior_skill):
    return confusing_factor[confusing_factor['item_asked_id'] == item_asked].\
        set_index('item_answered_id')['confusing_factor'].to_dict()

def adjusted_distractors(target_probability, item_asked, prior_skill):
    item_difficulty = sigmoid(prior_skill - difficulty[difficulty['item_id'] == item_asked]['difficulty'].values[0])
    level = get_level(target_probability, item_difficulty)
    weights = sorted(competive_distractors(target_probability, item_asked, prior_skill).items(), key=lambda x: x[1])
    weights = zip(weights, zip(*weights)[1][::-1])
    weight_median = np.median(zip(*weights)[1])
    return dict([(i, adjust_to_level(level, w, w_op, weight_median)) for ((i, w), w_op) in weights])


def plot_distractors_items(target_probability, prior_skill, term_name):
    item_asked = flashcards[flashcards['term_name'] == term_name]['item_id'].values[0]
    item_difficulty = sigmoid(prior_skill - difficulty[difficulty['item_id'] == item_asked]['difficulty'].values[0])
    rcParams['figure.figsize'] = 15, 5
    order = competive_distractors(target_probability, item_asked, prior_skill)
    for i, strategy in enumerate([competive_distractors, adjusted_distractors]):
        plt.suptitle('{}, {}%'.format(term_name, int(item_difficulty * 100)))
        plt.subplot(1, 2, i)
        plt.title(strategy.__name__)
        items, weights = zip(*sorted(strategy(target_probability, item_asked, prior_skill).items(), key=lambda (i, s): - order[i])[:20])
        flashcard_names = flashcards.set_index('item_id')['term_name'].to_dict()
        plt.plot(weights)
        plt.fill_between(range(len(weights)), [0] * len(weights), weights, alpha=0.2)
        plt.xlim(0, len(weights) - 1)
        plt.ylim(0, max(weights))
        plt.xticks(range(len(items)), map(lambda i: flashcard_names[i], items), rotation=90)
    plt.tight_layout()

def demo_distractors():
    term_name = widgets.SelectWidget(values=sorted(flashcards['term_name'].unique()))
    target_probability = widgets.FloatSliderWidget(min=0, max=1, step=0.01, value=0.75)
    prior_skill = widgets.FloatSliderWidget(min=-3, max=3, step=0.1, value=0)
    w = widgets.interactive(
        plot_distractors_items,
        target_probability=target_probability,
        term_name=term_name,
        prior_skill=prior_skill
    )
    display(w)    
demo_distractors()
```
